```python
import os
import itertools
```

## Create positive pairs

```python
images = os.listdir('Data')
```

```python
image_groups = {}
for image in images:
    person_id = image.split('_')[0]
    if person_id not in image_groups:
        image_groups[person_id] = []
    image_groups[person_id].append(image)
image_groups
```

```python
len(image_groups.keys())
```

```python
image_groups["Amrit"]
```

```python
positive_pairs = []
for person_id, images in image_groups.items():
    # Generate all possible pairs of images for the current person
    pairs = list(itertools.combinations(images, 2))
    # Add pairs to the list of positive pairs
    positive_pairs.extend(pairs)
    # positive_pairs.extend([(pair,1) for pair in pairs])
```

```python
len(positive_pairs)
```

```python
import random
```

```python
negative_pairs = []
num_positive_pairs = len(positive_pairs)

# Loop through positive pairs
for i in range(num_positive_pairs):
    # Randomly select a positive pair
    positive_pair = random.choice(positive_pairs)
    
    # Get the person IDs in the positive pair
    person_id_1 = positive_pair[0].split('_')[0]
    person_id_2 = positive_pair[1].split('_')[0]
    
    # Randomly select a person ID different from the person IDs in the positive pair
    candidate_person_ids = [person_id for person_id in image_groups.keys() if person_id != person_id_1 and person_id != person_id_2]
    negative_person_id = random.choice(candidate_person_ids)
    
    # Randomly select an image from the chosen person ID
    negative_image = random.choice(image_groups[negative_person_id])
    
    # Add the negative pair to the list
    negative_pairs.append((positive_pair[0], negative_image))
```
